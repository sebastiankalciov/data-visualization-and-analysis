As part of preproccesing for handling missing values:
- checking if there exist missing values
- if missing values do exist, print the process to know exactly what happens and then fill the voids with predefined strings

For handling duplicate rows:
- check for duplicate rows
- remove any duplicate rows

```python
import numpy as np

for df_name, df in dataframes.items():
    print(f"\n preprocessing dataframe: {df_name}")

    # handle missing values
    print(f"checking for missing values in '{df_name}'")
    missing_before = df.isnull().sum()
    missing_before = missing_before[missing_before > 0]

    if not missing_before.empty:
        print("missing values before handling:")
        print(missing_before)

        # specific handling for known missing columns identified from df.info()
        if 'tester' in df.columns and df['tester'].isnull().any():
            print("filling missing 'tester' values with 'Unknown'")
            df['tester'].fillna('Unknown', inplace=True)

        if 'task_description_plain_eng' in df.columns and df['task_description_plain_eng'].isnull().any():
            print("filling missing 'task_description_plain_eng' values with an empty string")
            df['task_description_plain_eng'].fillna('', inplace=True)

        # for 'solution' columns, check for empty strings and fill
        if 'solution' in df.columns:
            # replace empty strings (or strings with only whitespace) with NaN first
            df['solution'] = df['solution'].replace(r'^\s*$', np.nan, regex=True)

            if df['solution'].isnull().any():
                 print("filling empty 'solution' strings with '[EMPTY_SOLUTION]'")
                 df['solution'].fillna('[EMPTY_SOLUTION]', inplace=True)

        missing_after = df.isnull().sum()
        missing_after = missing_after[missing_after > 0]

        if not missing_after.empty:
            print("missing values after handling (still remaining):")
            print(missing_after)
        else:
            print("all identified missing values have been handled")
    else:
        print("no missing values found")

    # handle duplicate rows
    print(f"checking for duplicate rows in '{df_name}'")
    initial_rows = df.shape[0]
    df.drop_duplicates(inplace=True)
    rows_after_dropping = df.shape[0]

    if initial_rows > rows_after_dropping:
        print(f"removed {initial_rows - rows_after_dropping} duplicate rows")
    else:
        print("no duplicate rows found")

    # update the dataframe in the dict
    dataframes[df_name] = df
    print("\n" + "="*150)
```

Part of preprocessing is the normalization of code
Because the dataset contains only snippets of code, the normalization of code is necessary to have an ideal version of each snippet of code.
Therefore, to normalize the code:
- remove useless whitespaces
- remove empty lines
- join lines with \n to remove any extra spaces
- replace spaces or tabs with a single space

```python
import re

def normalize_code(code):
    if not isinstance(code, str):
        return code

    # remove unnecessary whitesapces from the end of each line
    lines = [line.strip() for line in code.splitlines()]

    # filter out empty lines
    lines = [line for line in lines if line]

    # join lines back with a single newline and remove any extra spaces
    normalized_code = '\n'.join(lines)

    # replace multiple spaces/tabs with a single space
    normalized_code = re.sub(r'[ \t]+', ' ', normalized_code)

    # remove unnecessary whitesapces from the whole block
    normalized_code = normalized_code.strip()
    return normalized_code

for df_name, df in dataframes.items():
    print(f"\ndataframe: {df_name}")
    if 'skeleton' in df.columns:
        print("skeleton column")
        df['skeleton'] = df['skeleton'].apply(normalize_code)
    if 'answer' in df.columns:
        print("answer column")
        df['answer'] = df['answer'].apply(normalize_code)
    if 'solution' in df.columns:
        print("solution column")
        df['solution'] = df['solution'].apply(normalize_code)
    dataframes[df_name] = df
    print("="*150)
```

For further analysis, we need each solution dataframe combined into a single dataframe

```python
all_solutions_df_list = []

for df_name, df in dataframes.items():
    if 'solutions' in df_name:
        year = int(df_name.split('-')[1]) # extract year from 'year-X-solutions'
        temp_df = df.copy() # create a copy to avoid modifying original dataframe
        temp_df['year'] = year
        all_solutions_df_list.append(temp_df)

# concatenate all solution dataframes
master_solutions_df = pd.concat(all_solutions_df_list, ignore_index=True)

print(f"shape master_solutions_df: {master_solutions_df.shape}")
display(master_solutions_df.head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import re

# a func to convert code to lowercase and to tokenize using word characters
def code_tokenizer(code):
    if not isinstance(code, str):
        return []
    return re.findall(r'\b\w+\b', code.lower())

# init TfidfVectorizer with code tokenizer
tfidf_vectorizer = TfidfVectorizer(
    tokenizer=code_tokenizer,
    stop_words=None, # no predefined stop words, as code keywords are important
    token_pattern=None # code tokenizer is used, so token_pattern is ignored
)

# apply TF-IDF to the solution column
tfidf_matrix = tfidf_vectorizer.fit_transform(master_solutions_df['solution'])

print(f"shape TF-IDF matrix {tfidf_matrix.shape}")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# calculate the cosine similarity between the TF-IDF feature vectors
similarity_matrix = cosine_similarity(tfidf_matrix)

print(f"shape similarity_matrix {similarity_matrix.shape}")
```

```python
import numpy as np

# define a similarity threshold
similarity_threshold = 0.95

# list to store highly similar pairs
highly_similar_solutions = []

n_solutions = similarity_matrix.shape[0]

for i in range(n_solutions):
    # iterate only through the upper triangle to avoid duplicates and self-comparison
    for j in range(i + 1, n_solutions):
        if similarity_matrix[i, j] > similarity_threshold:
            # get metadata for solution i
            sol1_question_id = master_solutions_df.loc[i, 'question_id']
            sol1_student_id = master_solutions_df.loc[i, 'student_id']
            sol1_year = master_solutions_df.loc[i, 'year']

            # get metadata for solution j
            sol2_question_id = master_solutions_df.loc[j, 'question_id']
            sol2_student_id = master_solutions_df.loc[j, 'student_id']
            sol2_year = master_solutions_df.loc[j, 'year']

            highly_similar_solutions.append({
                'solution1_index': i,
                'solution2_index': j,
                'similarity_score': similarity_matrix[i, j],
                'solution1_question_id': sol1_question_id,
                'solution1_student_id': sol1_student_id,
                'solution1_year': sol1_year,
                'solution2_question_id': sol2_question_id,
                'solution2_student_id': sol2_student_id,
                'solution2_year': sol2_year
            })

# convert the list of dictionaries to a pandas DataFrame
plagiarism_pairs = pd.DataFrame(highly_similar_solutions)

print(f"{len(plagiarism_pairs)} pairs of solutions with similarity > {similarity_threshold}")
display(plagiarism_pairs.head())
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=42, n_init='auto')

# fit KMeans to the TF-IDF matrix to get cluster assignments
cluster_labels = kmeans.fit_predict(tfidf_matrix)

# store the cluster labels in a new column in master_solutions_df
master_solutions_df['cluster_label'] = cluster_labels

display(master_solutions_df.head())
```

```python
from sklearn.ensemble import IsolationForest

# contamination is the proportion of outliers in the data set; this is an estimate
iso_forest = IsolationForest(random_state=42, contamination=0.05)

# fit the model and predict outlier status (-1 for outliers, 1 for inliers)
outlier_labels = iso_forest.fit_predict(tfidf_matrix)

# store the outlier labels in master_solutions_df
master_solutions_df['is_outlier'] = outlier_labels

print(f"total solutions {len(master_solutions_df)}")
print(f"outliers {sum(master_solutions_df['is_outlier'] == -1)}")

display(master_solutions_df.head())
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

tsne = TSNE(n_components=2, random_state=42, perplexity=30, init='random')

# fit and transform the TF-IDF matrix to get 2D embeddings
tsne_embeddings = tsne.fit_transform(tfidf_matrix)

print(f"shape t-SNE embeddings: {tsne_embeddings.shape}")

# create a new df for plotting
plot_df = pd.DataFrame(tsne_embeddings, columns=['tsne_component_1', 'tsne_component_2'])
plot_df['cluster_label'] = master_solutions_df['cluster_label'].astype(str)
plot_df['is_outlier'] = master_solutions_df['is_outlier']

# create the scatter plot
plt.figure(figsize=(12, 10))
sns.scatterplot(
    x='tsne_component_1',
    y='tsne_component_2',
    hue='cluster_label',
    style='is_outlier',
    palette='viridis',
    data=plot_df,
    legend='full',
    alpha=0.7
)

plt.title('t-SNE Visualization of Solution Clusters and Outliers')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import json

def generate_network_data(master_solutions_df, plagiarism_pairs, question_id_to_title_map, nodes_filename='/content/gdrive/MyDrive/Colab Notebooks/DVA-Project/nodes.json', links_filename='/content/gdrive/MyDrive/Colab Notebooks/DVA-Project/links.json'):
    # generate nodes data
    nodes = []
    for idx, row in master_solutions_df.iterrows():
        nodes.append({
            "id": idx,
            "year": int(row['year']),
            "cluster": int(row['cluster_label']),
            "is_outlier": bool(row['is_outlier'] == -1),
            "loc": int(row['lines_of_code']),
            "question": question_id_to_title_map.get(row['question_id'], "Unknown Question")
        })

    # generate links data
    links = []
    for idx, row in plagiarism_pairs.iterrows():
        links.append({
            "source": int(row['solution1_index']),
            "target": int(row['solution2_index']),
            "similarity": float(row['similarity_score'])
        })

    # write to JSON files
    with open(nodes_filename, 'w') as f:
        json.dump(nodes, f, indent=2)

    with open(links_filename, 'w') as f:
        json.dump(links, f, indent=2)

# call the func to generate the files
generate_network_data(
    master_solutions_df,
    plagiarism_pairs,
    question_id_to_title_map
)
```
